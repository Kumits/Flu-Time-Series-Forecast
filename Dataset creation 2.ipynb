```python
import pandas as pd
import tabula
import numpy as np
import glob
import re
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 550)
dane = pd.read_csv('C:/Users/Laptop/Desktop/Grypa/Dane/dane.csv')
dane.drop(dane.columns[dane.columns.str.contains('unnamed', case = False)], axis=1, inplace=True)
dane['Rok_str'] = dane['Rok_str'].astype(str)
dane
```

```python
suma = dane['0 to 4'].sum()
suma
```

```python
#przygotowanie do zacignicia danych
dane['Rok_str'] = dane['Rok'].astype(str)
dane['Sezon'] = ""
dane['0 to 4'] = ""
dane['5 to 14'] = ""
dane['15 to 64'] = ""
dane['65+'] = ""
```

```python
#zacignicie danych
for x in range(507):
    ad = dane.loc[x, "Plik"]
    
    liczba = tabula.read_pdf(ad, area = (230, 240, 240, 294), pages=1)
    liczba_str = str(liczba)
    O_to_4 = re.sub('[^0-9]', '', liczba_str)
    dane.loc[x, "0 to 4"] = O_to_4
    print(O_to_4)
    
    liczba2 = tabula.read_pdf(ad, area = (230, 330, 240, 360), pages=1)
    liczba2_str = str(liczba2)
    piec_to_14 = re.sub('[^0-9]', '', liczba2_str)
    dane.loc[x, "5 to 14"] = piec_to_14
    print(piec_to_14)
    
    liczba3 = tabula.read_pdf(ad, area = (230, 400, 240, 440), pages=1)
    liczba3_str = str(liczba3)
    pietnascie_to_64 = re.sub('[^0-9]', '', liczba3_str)
    dane.loc[x, "15 to 64"] = pietnascie_to_64
    print(pietnascie_to_64)
    
    liczba4 = tabula.read_pdf(ad, area = (230, 485, 240, 530), pages=1)
    liczba4_str = str(liczba4)
    szescpiec = re.sub('[^0-9]', '', liczba4_str)
    dane.loc[x, '65+'] = szescpiec
    print(szescpiec)
    
dane.to_csv('C:/Users/Laptop/Desktop/Grypa/Dane/dane.csv')
```

```python
#przygotowanie do wprawadzenia sezonw
lista_lat = dane['Rok'].unique()
lista_lat_str = dane['Rok_str'].unique()
slow1 = dict(zip(lista_lat_str, lista_lat))
print(slow1)
```

```python
lista_sezonow = []
for x in lista_lat_str:
    nazwa_sezonu = 'sezon_' + x
    lista_sezonow.append(nazwa_sezonu)
    #print(nazwa_sezonu)
lista_sezonow
```

```python
#podzia danych na pliki z sezonami
for x in lista_lat_str:
    sez1 = dane.loc[(dane['Rok'] == slow1[x]) & (dane['Miesiac'] > 8)]
    sez2 = dane.loc[(dane['Rok'] == slow1[x] + 1) & (dane['Miesiac'] < 9)]
    sezon = sez1.append(sez2)
    #sezon.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
    #sezon.drop(["a"], axis=1, inplace=True)
    sezon1 = sezon.reset_index(drop=True)
    sezon1['Sezon'] = x
    sezon1.to_csv('C:/Users/Laptop/Desktop/Grypa/Dane/Sezony/Sezon_' + x + '.csv')
    #print(sezon1)
```

```python
#scalanie sezonw w jeden plik
path = r'C:/Users/Laptop/Desktop/Grypa/Dane/Sezony'
all_files = glob.glob(path + "/*.csv")
li = []
for x in all_files:
    plik = pd.read_csv(x, index_col=None, header=0)
    li.append(plik)
total = pd.concat(li, axis=0, ignore_index=True)
total.drop(total.columns[total.columns.str.contains('unnamed', case = False)], axis=1, inplace=True)
total['Rok_str'] = total['Rok_str'].astype(str)
total.to_csv('C:/Users/Laptop/Desktop/Grypa/Dane/' + 'total' +'.csv')
```
